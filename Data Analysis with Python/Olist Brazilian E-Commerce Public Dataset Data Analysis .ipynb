## Data Collection 

```python
import os
import pandas as pd
import numpy as np

pd.set_option("display.max_columns",None)
```

## Load Dataset

```python
#Set Dataset Path
path = r"C:\Users\linga\OneDrive\Desktop\Dataset"
#Load All CSV Files
customers = pd.read_csv(os.path.join(path, "olist_customers_dataset.csv"))

orders = pd.read_csv(os.path.join(path, "olist_orders_dataset.csv"))

order_items = pd.read_csv(os.path.join(path, "olist_order_items_dataset.csv"))

payments = pd.read_csv(os.path.join(path, "olist_order_payments_dataset.csv"))

reviews = pd.read_csv(os.path.join(path, "olist_order_reviews_dataset.csv"))

products = pd.read_csv(os.path.join(path, "olist_products_dataset.csv"))

sellers = pd.read_csv(os.path.join(path, "olist_sellers_dataset.csv"))

translation = pd.read_csv(os.path.join(path, "product_category_name_translation.csv"))

geolocation = pd.read_csv(os.path.join(path, "olist_geolocation_dataset.csv"))
```

```python
#Inspect Data
datasets = {
    "customers":customers,
    "orders":orders,
    "order_items":order_items,
    "payments":payments,
    "reviews":reviews,
    "products":products,
    "translation":translation,
    "sellers":sellers,
    "geo":geolocation
}

for name,df in datasets.items():
    print("-"*80)
    print(name)
    print(df.shape)
    print(df.head())
```

## Data Cleaning

```python
#Check Missing Values
for name,df in datasets.items():
    print(name)
    print(df.isnull().sum())
```

## Order Table

```python
orders.isnull().sum()
```

## Reviews Table

```python
#review_comment_title
reviews["review_comment_title"] = reviews["review_comment_title"].fillna("No Title")

#review_comment_message
reviews["review_comment_message"] = reviews["review_comment_message"].fillna("No Review")

#Check
reviews.isnull().sum()
```

## Products Table

```python
#Category
products["product_category_name"] = products["product_category_name"].fillna("Unknown")

#Numeric Columns 
#Use Median
cols = [
    "product_name_lenght",
    "product_description_lenght",
    "product_photos_qty",
    "product_weight_g",
    "product_length_cm",
    "product_height_cm",
    "product_width_cm"
]
for col in cols:
    products[col] = products[col].fillna(products[col].median())

#Check
products.isnull().sum()
```

## Final Verification

```python
datasets = {
    "customers": customers,
    "orders": orders,
    "order_items": order_items,
    "payments": payments,
    "reviews": reviews,
    "products": products,
    "translation": translation,
    "sellers": sellers,
    "geo": geolocation
}

for name, df in datasets.items():
    print(f"\n{name}")
    print(df.isnull().sum())
```

### **Note:** In Orders Table missing values of order_approved_at, order_delivered_carrier_date, and order_delivered_customer_date were intentionally retained because they represent legitimate business scenarios such as canceled, unavailable, or undelivered orders. Filling these values would distort delivery-time calculations and KPI analysis.

```python
#Check Duplicate Records
for name,df in datasets.items():
    print(name)
    print(df.duplicated().sum())
```

```python
#Remove Duplicates
geolocation.drop_duplicates(inplace=True)

#Verificat Duplicate Records
for name,df in datasets.items():
    print(name)
    print(df.duplicated().sum())
```

```python
#Convert Date Columns
date_cols = [

'order_purchase_timestamp',
'order_approved_at',
'order_delivered_carrier_date',
'order_delivered_customer_date',
'order_estimated_delivery_date'

]

for col in date_cols:
    orders[col]=pd.to_datetime(orders[col])
```

## Data Wrangling

```python
#Merge Category Translation
products = products.merge(
translation,
on="product_category_name",
how="left"
)
```

```python
#Merge Orders
df = orders.merge(customers,on="customer_id")

#Merge Order Items
df = df.merge(order_items,on="order_id")

#Merge Payments
df = df.merge(payments,on="order_id")

#Merge Reviews
df = df.merge(reviews,on="order_id")

#Merge Products
df = df.merge(products,on="product_id")

#Merge Sellers
df = df.merge(sellers,on="seller_id")

#Final Dataset
print(df.shape)

df.head()
```

## Creating Required Features


```python
#Total Revenue
df["Revenue"] = df["price"] + df["freight_value"]

#Delivery Days
df["Delivery_Days"] = (
df["order_delivered_customer_date"]-
df["order_purchase_timestamp"]
).dt.days

#Purchase Month
df["Month"] = df["order_purchase_timestamp"].dt.month_name()

#Purchase Year
df["Year"] = df["order_purchase_timestamp"].dt.year

#Purchase Quarter
df["Quarter"] = df["order_purchase_timestamp"].dt.quarter

#Weekday
df["Weekday"] = df["order_purchase_timestamp"].dt.day_name()

#Delivery Status
df["Delivery_Status"] = np.where(
df["order_delivered_customer_date"]<=df["order_estimated_delivery_date"],
"On Time",
"Delayed"
)
```

## KPIs for Python

```python

#Total Revenue
total_revenue = df["Revenue"].sum()

print(f"Total Revenue : {total_revenue:,.2f}")

#Total Orders
total_orders = df["order_id"].nunique()

print(f"Total Orders : {total_orders}")

#Total Customers
total_customers = df["customer_unique_id"].nunique()

print(f"Total Customers : {total_customers}")

#Average Order Value (AOV)
#Since one order can contain multiple products, calculate revenue at the order level first.
order_revenue = (
    df.groupby("order_id")["Revenue"]
      .sum()
)
average_order_value = order_revenue.mean()

print(f"Average Order Value : {average_order_value:.2f}")

#Average Review Score
average_review_score = (
    df.groupby("order_id")["review_score"]
      .first()  #Using .first() avoids counting the same review multiple times after merging.
      .mean()
)

print(f"Average Review Score : {average_review_score:.2f}")

#Average Delivery Days
#Only delivered orders should be considered.
average_delivery_days = (
    df["Delivery_Days"]
    .dropna()
    .mean()
)

print(f"Average Delivery Days : {average_delivery_days:.2f}")


#Monthly Revenue
print("-" * 30)
print("Monthly Revenue")
print("-" * 30)
monthly_revenue = (
    df.groupby("Month")["Revenue"]
      .sum()
      .reset_index()
)
monthly_revenue["Month"] = monthly_revenue["Month"].astype(str)

print(monthly_revenue)

#Top Product Categories
print("-" * 30)
print('Product Categories')
print("-" * 30)
top_categories = (
    df.groupby("product_category_name_english")["Revenue"]
      .sum()
      .sort_values(ascending=False)
      .reset_index()
)

print(top_categories.head(10))

#Top Sellers
print("-" * 30)
print('Top Sellers')
print("-" * 30)
top_sellers = (
    df.groupby("seller_id")["Revenue"]
      .sum()
      .sort_values(ascending=False)
      .reset_index()
)

print(top_sellers.head(10))

#Payment Method Distribution
print("-" * 30)
print('Payment Method Distribution')
print("-" * 30)
payment_distribution = (
    df.groupby("payment_type")
      .agg(
          Total_Transactions=("order_id", "count"),
          Total_Revenue=("payment_value", "sum")
      )
      .sort_values("Total_Transactions", ascending=False)
      .reset_index()
)

print(payment_distribution)

```
